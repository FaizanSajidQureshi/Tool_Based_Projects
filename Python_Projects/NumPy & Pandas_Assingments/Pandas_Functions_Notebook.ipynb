
# Introduction to Pandas

Pandas is a powerful Python library for data manipulation and analysis. It provides two main data structures:
- **Series**: 1-dimensional labeled array.
- **DataFrame**: 2-dimensional labeled data structure with columns of potentially different types.

In this notebook, we will explore key functions in Pandas that are essential for data analysis.
    

```python
import pandas as pd
```


## 1. Creating a DataFrame

You can create a `DataFrame` from various data structures, such as dictionaries or lists.

### Example
    

```python

# Creating DataFrame from a dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'Occupation': ['Engineer', 'Doctor', 'Artist']}

df = pd.DataFrame(data)
df
    
```


## 2. Viewing Data in a DataFrame

Pandas provides functions to quickly view data in a `DataFrame`:
- `.head()`: Returns the first 5 rows.
- `.tail()`: Returns the last 5 rows.
- `.info()`: Provides a concise summary of the DataFrame.
- `.describe()`: Provides statistical summary for numerical columns.

### Example
    

```python

# Viewing the first few rows of data
df.head()

# Viewing the last few rows of data
df.tail()

# Viewing summary information
df.info()

# Statistical summary of numerical columns
df.describe()
    
```


## 3. Selecting Data

To select data from a DataFrame:
- `.loc[]`: Select rows and columns by labels.
- `.iloc[]`: Select rows and columns by position.

### Example
    

```python

# Select specific rows and columns using .loc
df_loc = df.loc[0:1, ['Name', 'Occupation']]

# Select specific rows and columns using .iloc (by position)
df_iloc = df.iloc[0:2, [0, 2]]

df_loc, df_iloc
    
```


## 4. Adding New Columns

You can add new columns to a DataFrame by assigning values.

### Example
    

```python

# Adding a new column to the DataFrame
df['Salary'] = [70000, 80000, 50000]
df
    
```


## 5. Dropping Columns or Rows

To drop columns or rows:
- `.drop()`: Remove columns or rows by labels.

### Example
    

```python

# Dropping the 'Salary' column
df_dropped = df.drop(columns=['Salary'])
df_dropped
    
```


## 6. Sorting Data

You can sort data in a DataFrame using:
- `.sort_values()`: Sort by the values of one or more columns.
- `.sort_index()`: Sort by the index.

### Example
    

```python

# Sorting by 'Age'
df_sorted = df.sort_values(by='Age')
df_sorted
    
```


## 7. Handling Missing Data

Pandas provides functions to handle missing data:
- `.isnull()`: Detect missing values.
- `.dropna()`: Remove missing data.
- `.fillna()`: Fill missing data with specified values.

### Example
    

```python

# Adding NaN values for demonstration
import numpy as np
df_missing = df.copy()
df_missing.loc[1, 'Age'] = np.nan

# Detecting missing values
df_missing.isnull()

# Filling missing values with a specified value
df_filled = df_missing.fillna(0)

# Dropping rows with missing values
df_dropped_na = df_missing.dropna()

df_filled, df_dropped_na
    
```


## 8. Grouping Data

The `.groupby()` function is used to group data by one or more columns, and then apply functions to each group.

### Example
    

```python

# Grouping by 'Occupation' and calculating the mean of 'Age'
df_grouped = df.groupby('Occupation')['Age'].mean()
df_grouped
    
```


## 9. Merging and Joining

Pandas provides functions to merge or join DataFrames:
- `.merge()`: Merge DataFrames based on common columns.
- `.join()`: Join DataFrames by index.

### Example
    

```python

# Creating another DataFrame to merge
data2 = {'Name': ['Alice', 'Bob', 'Charlie'],
         'Department': ['HR', 'Finance', 'Marketing']}

df2 = pd.DataFrame(data2)

# Merging the two DataFrames
df_merged = pd.merge(df, df2, on='Name')
df_merged
    
```

## 10. Filtering

```python
df[df['Department']=='HR']

df.loc[df['Age']>50, ['Name', 'Salary', 'Age']]

df[(df['Age']>50) | (df['Department']=='Finance')]

df[df['Department'].isin(['Sales','Finance'])]
```

# Matplotlib Examples

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Line plot
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='sin(x)')
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()

# Scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(x, y, color='red')
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()

# Bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 4]
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color='green')
plt.title('Bar Plot')
plt.ylabel('Value')
plt.show()

# Histogram
data = np.random.randn(1000)
plt.figure(figsize=(6, 4))
plt.hist(data, bins=30, color='purple')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Box plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.figure(figsize=(6, 4))
plt.boxplot(data, vert=True)
plt.title('Box Plot')
plt.xlabel('Groups')
plt.ylabel('Values')
plt.show()

# Pie chart
sizes = [25, 35, 20, 20]
labels = ['Apple', 'Banana', 'Cherry', 'Date']
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.show()
```

| Violin plot  | `sns.violinplot(x, y, data=df)`  | Density + box plot combined |
| Heatmap      | `sns.heatmap(data)`              | Matrix-like colored plot    |
| Pair plot    | `sns.pairplot(df)`               | Pairwise scatter plots      |


# Seaborn Example

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create example DataFrame
df = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C'], size=100),
    'Value1': np.random.randn(100),
    'Value2': np.random.randn(100)
})

# Scatter plot
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Value1', y='Value2', hue='Category', data=df)
plt.title('Seaborn Scatter Plot')
plt.show()

# Line plot
df_line = pd.DataFrame({
    'x': np.arange(10),
    'y': np.random.randn(10)
})
plt.figure(figsize=(6, 4))
sns.lineplot(x='x', y='y', data=df_line)
plt.title('Seaborn Line Plot')
plt.show()

# Bar plot
df_bar = df.groupby('Category')['Value1'].mean().reset_index()
plt.figure(figsize=(6, 4))
sns.barplot(x='Category', y='Value1', data=df_bar)
plt.title('Seaborn Bar Plot')
plt.show()

# Box plot
plt.figure(figsize=(6, 4))
sns.boxplot(x='Category', y='Value1', data=df)
plt.title('Seaborn Box Plot')
plt.show()

# Violin plot
plt.figure(figsize=(6, 4))
sns.violinplot(x='Category', y='Value1', data=df)
plt.title('Seaborn Violin Plot')
plt.show()

# Heatmap
corr = df[['Value1', 'Value2']].corr()
plt.figure(figsize=(5, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Seaborn Heatmap')
plt.show()

# Pair plot
sns.pairplot(df, hue='Category')
plt.suptitle('Seaborn Pair Plot', y=1.02)
plt.show()
```
